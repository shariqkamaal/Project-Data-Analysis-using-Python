```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('D:\MCA course\DA internship\householdtask3.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Display the column names to identify the correct ones
print(df.columns)

# Bar Chart: Example for video views by category
df.groupby('year')['expenditure'].sum().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Total Expenditures')
plt.title('Total Expenditure over the Year')
plt.legend(['Expenses'])
plt.show()

# Line Chart: Example for video likes over time
df.plot(x='year', y='expenditure', kind='line')
plt.xlabel('Year')
plt.ylabel('Total Expenditures')
plt.title('Total Expenditure over the Year')
plt.legend(['Expenses'])
plt.show()
```
