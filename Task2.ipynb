```python
import pandas as pd

# Load CSV file into a DataFrame
def load_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        print("CSV file loaded successfully.")
        return df
    except FileNotFoundError:
        print("The specified file was not found.")
        return None

# Filter data based on conditions
def filter_data(df, condition):
    filtered_df = df.query(condition)
    return filtered_df

# Handle missing values
def handle_missing_values(df, method='drop', fill_value=None):
    if method == 'drop':
        df = df.dropna()
    elif method == 'fill':
        df = df.fillna(fill_value)
    return df

# Calculate summary statistics
def summary_statistics(df):
    return df.describe()

# Main function to demonstrate the operations
def main():
    # Path to your CSV file
    file_path = 'D:/MCA course/DA internship/Data Cleaning and Preprocessing.csv'
    
    # Load the CSV file
    df = load_csv(file_path)
    if df is None:
        return

    # Display the first few rows of the DataFrame
    print("Initial DataFrame:")
    print(df.head())
    
    # Display the column names to identify the correct ones
    print("Column Names:", df.columns)
    
    
    condition = 'BlowFlow > 1200'  
    filtered_df = filter_data(df, condition)
    print("\nFiltered DataFrame (BlowFlow > 1200):")
    print(filtered_df.head())
    
    # Handle missing values by dropping rows with any missing values
    df_no_missing = handle_missing_values(df, method='drop')
    print("\nDataFrame after dropping missing values:")
    print(df_no_missing.head())
    
    # Handle missing values by filling with a specific value (example: filling with 0)
    df_filled = handle_missing_values(df, method='fill', fill_value=0)
    print("\nDataFrame after filling missing values with 0:")
    print(df_filled.head())
    
    # Calculate and display summary statistics
    stats = summary_statistics(df)
    print("\nSummary Statistics:")
    print(stats)

# Run the main function
if __name__ == "__main__":
    main()
```
